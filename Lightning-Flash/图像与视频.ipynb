# 任务

识别图像中的内容的任务称为图像分类。 通常，图像分类用于识别包含单个对象的图像。 该任务在一定程度上预测图像最有可能属于哪个“类别”。 类是描述图像中内容的标签，例如“汽车”、“房子”、“猫”等。

# 例子

让我们看一下使用膜翅目数据集预测图像是否包含蚂蚁或蜜蜂的任务。 数据集包含训练和验证文件夹，然后每个文件夹包含一个蜜蜂文件夹，其中包含蜜蜂的图片，以及一个包含蚂蚁图像的 ants 文件夹，您猜对了，是蚂蚁。

```python
hymenoptera_data
├── train
│   ├── ants
│   │   ├── 0013035.jpg
│   │   ├── 1030023514_aad5c608f9.jpg
│   │   ...
│   └── bees
│       ├── 1092977343_cb42b38d62.jpg
│       ├── 1093831624_fb5fbe2308.jpg
│       ...
└── val
    ├── ants
    │   ├── 10308379_1b6c72e180.jpg
    │   ├── 1053149811_f62a3410d3.jpg
    │   ...
    └── bees
        ├── 1032546534_06907fe3b3.jpg
        ├── 10870992_eebeeb3a12.jpg
        ...
```

使用 download_data() 下载数据后，我们创建 ImageClassificationData。 我们选择一个预训练的主干用于我们的 ImageClassifier 并对膜翅目数据进行微调。 然后我们使用经过训练的 ImageClassifier 进行推理。 最后，我们保存模型。 这是完整的示例：

```python
import torch

import flash
from flash.core.data.utils import download_data
from flash.image import ImageClassificationData, ImageClassifier

# 1. 创建数据模块
download_data("https://pl-flash-data.s3.amazonaws.com/hymenoptera_data.zip", "./data")

datamodule = ImageClassificationData.from_folders(
    train_folder="data/hymenoptera_data/train/",
    val_folder="data/hymenoptera_data/val/",
)

# 2. 构建任务
model = ImageClassifier(backbone="resnet18", num_classes=datamodule.num_classes)

# 3. 创建训练器并微调模型
trainer = flash.Trainer(max_epochs=3, gpus=1)
trainer.finetune(model, datamodule=datamodule, strategy="freeze")

# 4. 预测几张图片上的内容！ 蚂蚁还是蜜蜂？
predictions = model.predict(
    [
        "data/hymenoptera_data/val/bees/65038344_52a45d090d.jpg",
        "data/hymenoptera_data/val/bees/590318879_68cf112861.jpg",
        "data/hymenoptera_data/val/ants/540543309_ddbb193ee5.jpg",
    ]
)
print(predictions)

# 5. 保存模型！
trainer.save_checkpoint("image_classification_model.pt")
```

# 自定义转换

Flash 会自动应用一些默认的图像转换和增强，但您可能希望为自己的用例自定义这些。 基础预处理为数据加载管道中的不同阶段定义了 7 个钩子。 要应用图像增强，您可以直接从 flash.image.classification.transforms 导入 default_transforms，然后使用 merge_transforms() 辅助函数将自定义图像转换与它们合并。 这是我们加载默认转换并与自定义 torchvision 转换合并的示例。 我们使用 post_tensor_transform 钩子在图像转换为 torch.Tensor 后应用转换。

```python
from torchvision import transforms as T

import flash
from flash.core.data.data_source import DefaultDataKeys
from flash.core.data.transforms import ApplyToKeys, merge_transforms
from flash.image import ImageClassificationData, ImageClassifier
from flash.image.classification.transforms import default_transforms

post_tensor_transform = ApplyToKeys(
    DefaultDataKeys.INPUT,
    T.Compose([T.RandomHorizontalFlip(), T.ColorJitter(), T.RandomAutocontrast(), T.RandomPerspective()]),
)

new_transforms = merge_transforms(default_transforms((64, 64)), 
                                  {"post_tensor_transform": post_tensor_transform})

datamodule = ImageClassificationData.from_folders(
    train_folder="data/hymenoptera_data/train/", 
    val_folder="data/hymenoptera_data/val/", 
    train_transform=new_transforms
)

model = ImageClassifier(backbone="resnet18", 
                        num_classes=datamodule.num_classes)

trainer = flash.Trainer(max_epochs=1)
trainer.finetune(model, datamodule=datamodule, strategy="freeze")
```
